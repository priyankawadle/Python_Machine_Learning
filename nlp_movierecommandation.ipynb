```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(
    credits,
    how="inner",
    left_on="id",
    right_on="movie_id"
)
```

```python
movies = movies[['movie_id','title_y','overview','genres','keywords','cast','crew']]
```

```python
movies.sample(5)
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.reset_index(drop=True, inplace=True)
```

```python
movies.rename(columns={'title_y':'title'}, inplace=True)
```

```python
movies.duplicated().sum()   
```

```python
movies.head(5)
```

```python
import ast
```

```python
movies['genres']=movies['genres'].apply(lambda b:[x['name'] for x in ast.literal_eval(b)])
```

```python
movies['keywords']=movies['keywords'].apply(lambda b:[x['name'] for x in ast.literal_eval(b)])
```

```python
# we are taking the names of the 4 most prominent aactors in the movie
def fetch_cast(obj):
    counter=0
    L=[]
    for i in ast.literal_eval(obj):
        if counter!=4:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
fetch_cast(movies.iloc[3].cast)
```

```python
movies['cast']=movies['cast'].apply(fetch_cast)
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
fetch_director(movies.iloc[0].crew)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
movies['overview']=movies['overview'].str.split()
```

```python
movies['genres']=movies['genres'].apply(lambda j:[i.replace(" ","") for i in j])
movies['keywords']=movies['keywords'].apply(lambda j:[i.replace(" ","") for i in j])
movies['cast']=movies['cast'].apply(lambda j:[i.replace(" ","") for i in j])
movies['crew']=movies['crew'].apply(lambda j:[i.replace(" ","") for i in j])
```

```python
movies['tags']=movies['genres']+movies['cast']+movies['crew']+movies['keywords']+movies['overview']
```
