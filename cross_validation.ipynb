```python
import kagglehub

# Download latest version
path =  kagglehub.dataset_download("sakshisatre/titanic-dataset")

print("Path to dataset files:", path)
```

```python
import os 
import pandas as pd
os.listdir(path)
```

```python

df = pd.read_csv(os.path.join(path,'Titanic Dataset.csv'))
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df['sex']=df['sex'].map({'female':0,'male':1})
```

```python
df
```

```python
age_mean = df['age'].mean()
```

```python
df['age'].fillna(age_mean, inplace=True)
```

```python
df["name"] = df["name"].astype("category")
df["sex"] = df["sex"].astype("category")
df["ticket"] = df["ticket"].astype("category")
df["embarked"] = df["embarked"].astype("category")
df["cabin"] = df["cabin"].astype("category")
df["boat"] = df["boat"].astype("category")
df["home.dest"] = df["home.dest"].astype("category")
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['cabin','boat','body','home.dest'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=['embarked'])
```

```python
df[['pclass', 'age', 'sibsp', 'parch','fare']].corr()
```

```python
df.drop(["name","ticket"],axis=1,inplace=True)
```

```python
traning_cols = ['pclass', 'sex', 'age', 'sibsp', 'parch',
       'fare', 'embarked_C','embarked_Q','embarked_S']
testing_cols = ['survived']
X = df[traning_cols]
y = df[testing_cols]
```

```python
#corelation

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
corr=df[['pclass', 'age', 'sibsp', 'parch','fare']].corr().round(2)
sns.heatmap(corr,annot=True)
```
